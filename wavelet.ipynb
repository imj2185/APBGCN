```python
import torch
import torch.nn
from pytorch_wavelets import DWT1DForward, DWT1DInverse
import matplotlib.pyplot as plt
from torch_geometric.data import DataLoader
from data.dataset3 import SkeletonDataset
```

```python
ds = SkeletonDataset(root="/home/dusko/Documents/projects/APBGCN", name='ntu_60',
                                    benchmark='xsub', sample='val')

loader = DataLoader(ds[0:200], batch_size=1, shuffle=True)
```

```python
dwt = DWT1DForward(wave='db6', J=3)
#(F, 25, 3)
#(25,F,3)
yl_list = []
yh_list = []

for i, data in enumerate(loader):
    #print(data.y.item())
    if data.y.item() == 1:
        yl, yh = dwt(data.x.permute(1, 2, 0))
        yl_list.append(yl)
        yh_list.append(yh[0][0])
        print(yl.shape, yh[0].shape)
```

```python
plt.imshow(yh_list[0], cmap = 'coolwarm', aspect = 'auto')
plt.title(str(data.y))
plt.ylabel('Scale')
plt.xlabel('Time')
plt.tight_layout()
```

```python
plt.imshow(yh_list[1], cmap = 'coolwarm', aspect = 'auto')
plt.title(str(data.y))
plt.ylabel('Scale')
plt.xlabel('Time')
plt.tight_layout()
```
